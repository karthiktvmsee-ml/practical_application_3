### Problem 5: Engineering Features

Now that you understand your business objective, we will build a basic model to get started.  Before we can do this, we must work to encode the data.  Using just the bank information features, prepare the features and target column for modeling with appropriate encoding and transformations.

````python
# 1. Select bank client features and target
client_features = ["age", "job", "marital", "education",
                   "default", "housing", "loan"]
target_col = "y"

X_client = df[client_features].copy()
y = (df[target_col] == "yes").astype(int)   # yes -> 1, no -> 0

# 2. One-hot encode categorical bank client features
cat_cols = ["job", "marital", "education", "default", "housing", "loan"]

X_client_encoded = pd.get_dummies(X_client,
                                  columns=cat_cols,
                                  drop_first=False)  # keep all levels for now

print(X_client_encoded.head())
print(X_client_encoded.shape)
print(y.shape)
````

### Problem 6: Train/Test Split

With your data prepared, split it into a train and test set.

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_client_encoded,
    y,
    test_size=0.3,        # 30% test
    random_state=42,      # reproducible split
    stratify=y            # preserve class balance
)
````

### Problem 7: A Baseline Model

Before we build our first model, we want to establish a baseline.  What is the baseline performance that our classifier should aim to beat?

````python
'''The baseline is the performance of a trivial classifier\
that always predicts the majority class (always “no” for y).

In this dataset, about 11–12% of clients subscribe (“yes”)\
and 88–89% do not (“no”), so a classifier that always predicts\
“no” would achieve roughly 88–89% accuracy; the models should\
aim to beat this (and, importantly, achieve higher recall\
for the positive “yes” class)'''
````

### Problem 8: A Simple Model

Use Logistic Regression to build a basic model on your data.  

````python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(
    max_iter=1000,      # ensure convergence
    solver="lbfgs"      # default, works fine for this size
)

log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
y_proba = log_reg.predict_proba(X_test)[:, 1]
````

### Problem 9: Score the Model

What is the accuracy of your model?

````python
from sklearn.metrics import accuracy_score

test_accuracy = accuracy_score(y_test, y_pred)
print("Test accuracy:", test_accuracy)
````

### Problem 10: Model Comparisons

Now, we aim to compare the performance of the Logistic Regression model to our KNN algorithm, Decision Tree, and SVM models.  Using the default settings for each of the models, fit and score each.  Also, be sure to compare the fit time of each of the models.  Present your findings in a `DataFrame` similar to that below:

| Model | Train Time | Train Accuracy | Test Accuracy |
| ----- | ---------- | -------------  | -----------   |
|     |    |.     |.     |

````python
import time

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, solver="lbfgs"),
    "KNN": KNeighborsClassifier(),              # default k=5
    "Decision Tree": DecisionTreeClassifier(),  # default tree
    "SVM": SVC()                                # RBF kernel by default
}

results = []

for name, model in models.items():
    start = time.time()
    model.fit(X_train, y_train)
    end = time.time()

    train_pred = model.predict(X_train)
    test_pred  = model.predict(X_test)

    train_acc = accuracy_score(y_train, train_pred)
    test_acc  = accuracy_score(y_test, test_pred)
    train_time = end - start

    results.append({
        "Model": name,
        "Train Time": train_time,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc
    })

results_df = pd.DataFrame(results)
results_df
````

### Problem 11: Improving the Model

Now that we have some basic models on the board, we want to try to improve these.  Below, we list a few things to explore in this pursuit.


- Hyperparameter tuning and grid search.  All of our models have additional hyperparameters to tune and explore.  For example the number of neighbors in KNN or the maximum depth of a Decision Tree.  
- Adjust your performance metric

1. Example: hyperparameter tuning with GridSearchCV, for instance, for KNN and Decision Tree:

````python
from sklearn.model_selection import GridSearchCV

# KNN
knn = KNeighborsClassifier()
knn_param_grid = {
    "n_neighbors": [3, 5, 7,  9],
    "weights": ["uniform", "distance"]
}

knn_grid = GridSearchCV(
    knn,
    knn_param_grid,
    cv=5,
    scoring="f1",          # or 'recall' for positive class
    n_jobs=-1
)
knn_grid.fit(X_train, y_train)

print("Best KNN params:", knn_grid.best_params_)
print("Best KNN cv score:", knn_grid.best_score_)
````

````python
# Decision Tree
dt = DecisionTreeClassifier(random_state=42)
dt_param_grid = {
    "max_depth": [3, 5, 10, None],
    "min_samples_split": [2, 10, 50],
    "min_samples_leaf": [1, 5, 10]
}

dt_grid = GridSearchCV(
    dt,
    dt_param_grid,
    cv=5,
    scoring="f1",
    n_jobs=-1
)
dt_grid.fit(X_train, y_train)

print("Best DT params:", dt_grid.best_params_)
print("Best DT cv score:", dt_grid.best_score_)
````

2. Adjusting the performance metric:
 Because “yes” is the minority class, accuracy can look high even for a poor model that predicts all “no”. More suitable metrics include:

- Recall for the positive class (how many “yes” did we capture?).

- Precision (how many predicted “yes” were actually “yes”?).

- F1 score (harmonic mean of precision and recall).

- ROC AUC or PR AUC for ranking quality.

````python
from sklearn.metrics import classification_report, f1_score, roc_auc_score

print(classification_report(y_test, y_pred))  # shows precision/recall/F1
print("Test F1:", f1_score(y_test, y_pred))
print("Test ROC AUC:", roc_auc_score(y_test, y_proba))
````
